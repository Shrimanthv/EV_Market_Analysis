## The LMV(Light Motor Vehicle) which has a market share of 17.32%

The Light Motor Vehicle (LMV) category encompasses a range of vehicle platforms, including sedans, hatchbacks, SUVs, and MPVs (Multi-Purpose Vehicles).

```python
df_sorted = df.sort_values('Total Registration')

plt.figure(figsize=(10,6))
plt.plot(df_sorted['Vehicle Class'], df_sorted['Total Registration'], marker='o')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Vehicle Class')
plt.ylabel('Total Registration')
plt.title('Total Vehicle Registrations by Vehicle Class (Line Plot)')
plt.tight_layout()
plt.show()
```

## The analysis indicates that Two-Wheeler (Non-Transport) Two-Wheeler (Transport) vehicles hold the largest market share at 73.34%, followed by Light Motor

Based on the analysis, it is observed that the Two-Wheeler Non-Transport (NT) and Two-Wheeler Transport (NT) segment holds the largest share of the market, accounting for 73.34% of total vehicle registrations. This is followed by the Light Motor Vehicle (LMV) segment, which constitutes 17.32% of the market.

Given that Two-Wheeler Non-Transport (NT) and Transport (T) vehicles dominate the market with a 73.34% share, it would be strategically advantageous for the startup to initially focus on developing personal-use electric two-wheelers. This segment offers the largest customer base and the highest adoption potential. Additionally, considering that Light Motor Vehicles (LMVs) represent 17.32% of the market, the startup could also explore opportunities in the electric LMV segment (such as electric cars or small electric vans) as a secondary target for future expansion. Prioritizing two-wheelers aligns well with current market demand, faster adoption cycles, and relatively lower production and infrastructure costs compared to four-wheelers.
