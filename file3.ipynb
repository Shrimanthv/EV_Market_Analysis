## Visualising the vehicle type dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
## Load the dataset
df= pd.read_csv("indian-ev-data.csv")
```

```python
df.head()
```

```python
sns.countplot(data=df, x='Vehicle Type')
```

## From the above graph we can say that scootes sell more than bike 

```python
sns.scatterplot(data=df, x='Battery Capacity (kWh)', y='Range per Charge (km)',
                hue='Vehicle Type')
```

##The graph demonstrates a positive correlation between battery capacity and vehicle range, indicating that an increase in battery capacity is associated with an extended range.

```python
sns.histplot(df['Range per Charge (km)'], bins=10, kde=True,)
```

```python
df.columns
```

```python

plot_df = df[['Range per Charge (km)', 'Charging Time']]

# Sort values by Range for better visual order
plot_df = plot_df.sort_values(by='Range per Charge (km)')

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(plot_df['Range per Charge (km)'].astype(str), plot_df['Charging Time'], color='skyblue', edgecolor='black')

# Labels and title
plt.xlabel("Range per Charge (km)")
plt.ylabel("Charging Time (hours)")
plt.title("Charging Time vs. Range per Charge")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
## The graph clearly indicates that greater vehicle range is associated with increased charging time in most cases.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data
plot_df = df[[ 'Range per Charge (km)', 'Price']].dropna()
plot_df = plot_df.sort_values(by='Range per Charge (km)')

# Set the plot size and style
plt.figure(figsize=(14, 7))
sns.set(style="whitegrid")

# Create the bar plot
sns.barplot(
    data=plot_df,
    x='Range per Charge (km)',
    y='Price'
)

# Labels and title
plt.xlabel("Vehicle Model (sorted by range)")
plt.ylabel("Price (INR)")
plt.title("Price vs. Range per Charge")
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## The image also illustrates that an increase in vehicle range is generally accompanied by a higher vehicle cost.
